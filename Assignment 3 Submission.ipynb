```python
interest = df.groupby('Year')['luxury index'].mean().reset_index() #new dataframe with a grouped year and the mean of luxury index for each year

plt.bar(interest['Year'], interest['luxury index']) #plots bar graph
plt.xticks(interest['Year']) #sets the ticks to be the single year
plt.ylim(0, 50) #sets y-axis value range
plt.xlabel('Year') #labels x axis
plt.ylabel('Interest in Luxury Cosmetics') #labels y-axis
plt.title('Interest in Luxury Cosmetics from 2022-2025') #sets title
plt.show() #prints graph
```

The bar graph above visualizes the change in interest of luxury cosmetics over time, spanning from the years 2022-2025. From this graph, we see that there has been a semi-steady increase, with 2025 having the highest at almost 30. However, this change is in small increments which could mean the data is too similar to draw conclusions from. 

```python
quarterly = df.groupby(['Year', 'Period'])[['luxury index', 'Value']].mean().reset_index() #new dataframe grouped by year and period, taking the mean of luxury index and the value(ECI)

plt.plot(quarterly['Value'], quarterly['luxury index'], linestyle = 'none', marker = '*') #scatterplot for the quarterly ECI and luxury index
plt.xlabel('Wage Growth (ECI)') #labels x-axis
plt.ylabel('Interest in Luxury Cosmetics') #labels y-axis
plt.title('Interest in Luxury Cosmetics on Wage Growth') #sets title
plt.show() #shows graph
```

The graph above visualizes the relationship between wage growth and the interest in luxury cosmetics. As shown, there is a negative relationship as the highest interest in cosmetics is at the lowest point for wage growth, and vice versa. From this graph, we can conclude that the 'Lipstick Effect" has remained relevant through the 2020's. 

```python
fig, ax = plt.subplots(nrows = 1, ncols= 2, figsize=(12, 5))#creates figure named 'ax' with 1 row and 2 columns of graphs, sized 12x5

ax1 = plt.subplot(1, 2, 1) #sets first axis to be bar graph
ax1.bar(interest['Year'], interest['luxury index'])
ax1.set_xticks(interest['Year'])
ax1.set_ylim(0, 50)
ax1.set_xlabel('Year')
ax1.set_ylabel('Interest in Luxury Cosmetics')
ax1.set_title('Interest in Luxury Cosmetics from 2022-2025')

ax2 = plt.subplot (1, 2, 2) #sets second axis as scatterplot
ax2.plot(quarterly['Value'], quarterly['luxury index'], linestyle = 'none', marker = '*')
ax2.set_xlabel('Wage Growth (ECI)')
ax2.set_ylabel('Interest in Luxury Cosmetics')
ax2.set_title('Interest in Luxury Cosmetics on Wage Growth')

plt.show()
```

```python
df.boxplot(column='luxury index', by='Year') #creates boxplot showing distribution of data
plt.xlabel('Year') #labels x-axis
plt.ylabel('Interest in Luxury Cosmetics') #labels y-axis
plt.title('Distribution of Luxury Cosmetic Interest by Year') #sets title
plt.suptitle('') #removes automatic boxplot title
plt.show() #shows graph
```

The above box plot visualizes the distribution of luxury cosmetic interest by year, showing how it has increased overtime and any outliers in the data. From this graph, we see that interest in luxury cosmetics increases over time but is variable and uneven, meaning consumer interest has become less consistent and more volatile as years go on.
